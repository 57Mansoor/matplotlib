```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset (you can replace the path with your own dataset)
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Set the style of seaborn for a cleaner look
sns.set(style="whitegrid")

# 1. Bar plot: Distribution of survivors vs non-survivors
def plot_survival_distribution():
    plt.figure(figsize=(6, 4))
    sns.countplot(x='Survived', data=df, palette='Set2')
    plt.title('Survival Distribution (0 = No, 1 = Yes)')
    plt.xlabel('Survived')
    plt.ylabel('Count')
    plt.xticks(ticks=[0, 1], labels=["No", "Yes"])
    plt.show()

# 2. Scatter plot: Age vs Fare
def plot_age_vs_fare():
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x='Age', y='Fare', data=df, hue='Survived', palette='coolwarm', alpha=0.6)
    plt.title('Age vs Fare (colored by Survival)')
    plt.xlabel('Age')
    plt.ylabel('Fare')
    plt.show()

# 3. Heatmap: Correlation Matrix
def plot_correlation_heatmap():
    plt.figure(figsize=(10, 8))
    corr_matrix = df.corr()  # Get the correlation matrix for numeric columns
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()

# 4. Boxplot: Fare distribution by Survival
def plot_fare_by_survival():
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Survived', y='Fare', data=df, palette='Set2')
    plt.title('Fare Distribution by Survival')
    plt.xlabel('Survived')
    plt.ylabel('Fare')
    plt.xticks(ticks=[0, 1], labels=["No", "Yes"])
    plt.show()

# 5. Pairplot: Relationship between multiple features
def plot_pairplot():
    sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived', palette='coolwarm', height=3)
    plt.suptitle('Pairplot: Relationships Between Features', y=1.02)
    plt.show()

# Main function to execute all visualizations
def main():
    print("Visualizing Titanic dataset using Matplotlib and Seaborn")
    
    # Call the plotting functions
    plot_survival_distribution()
    plot_age_vs_fare()
    plot_correlation_heatmap()
    plot_fare_by_survival()
    plot_pairplot()

if __name__ == "__main__":
    main()
```
